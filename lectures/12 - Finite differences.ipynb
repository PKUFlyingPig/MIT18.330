```julia
hs = 0.001:0.001:0.5
```

```julia
f(x) = x^3 - 2x^2 + 1

a = 2.0
```

```julia
fp(x) = 3x^2 - 4x
```

```julia
fp(a)
```

```julia
g(h) = (f(a + h) - f(a)) / h
```

```julia
data = [(f(a + h) - f(a)) / h for h in hs];
```

```julia
using Plots
```

```julia
scatter(hs, data)
```

```julia
# range from 1e-16 to 1e-1 equally spaced on a log axis
```

```julia
hs = [10.0^(i) for i in -16:0.01:-1];
```

```julia
scatter(hs, g.(hs))
```

```julia
δs = g.(hs) .- 4;
```

```julia
scatter(hs, abs.(δs), scale=:log10)
plot!(x -> x)
```

```julia
δs
```

```julia
g2(h) = (f(a + h) - f(a - h)) / (2h) 
```

```julia
δ2s = g2.(hs) .- 4;

scatter(hs, abs.(δs), scale=:log10, alpha=0.5, label="forward")
scatter!(hs, abs.(δ2s), scale=:log10, alpha=0.5, label="centered")

plot!(h -> h, ls=:dash)
plot!(h -> h^2, ls=:dash, leg=:bottomright)
```
